1. Una clase es un un objeto estandarizado que describe propiedades y acciones comunes a un conjunto de entidades, un objeto muy general que se utiliza para crear mas objetos con propiedades similares. Mientras que, un objeto es una representación individual de una clase con características propias relacionadas con la clase.

2. Self sirve para hacer referencia al contenido del objeto con el que está asociado.

3. Los constructores se utilizan para crear instancias sobre los objetos de una clase. La función de los constructores es asignar valores a los elementos de clase cuando se crea un objeto de esa misma clase.


4. Los atributos de clase son aquellos que se definen dentro de la clase pero fuera de cualquier método. Estos atributos son compartidos por todas las instancias de la clase. Esto significa que, si se modifica el valor de un atributo de clase en una instancia, ese cambio se reflejará en todas las demás instancias de la clase.
Mientras que, los atributos de instancia son aquellos que se definen dentro del método __init__ de la clase. Cada instancia de la clase tiene sus propios valores para los atributos de instancia, lo que significa que los cambios en estos atributos solo afectan a esa instancia específica.

````python
class Carro():
    def __init__(self, marca, color):
        self.marca = marca
        self.color = color

    def encender(self):
        print('!Listo para conducir!')

class Camioneta(Carro):
    def __init__(self, marca, color, tamaño):
        super().__init__(marca, color)
        self.tamaño = tamaño

    def encender(self, llave):
        if llave == 'llave_camioneta':
            print('¡Listo para conducir!')
        else:
            print('La llave no es la correcta')

    def apagar(self, apagar):
        if apagar:
            print('¡El motor está apagado!')
        else:
            print('¡Sigo conduciendo!')
````

````python
class Ropa:
    def __init__(self, tipo, talla, color, precio):
        self.tipo = tipo
        self.talla = talla
        self.color = color
        self.precio = precio

    def info(self):
        print(f"Tipo: {self.tipo}")
        print(f"Talla: {self.talla}")
        print(f"Color: {self.color}")
        print(f"Precio: ${self.precio}")

    def descuento(self, descuento_porcentaje):
        descuento = self.precio * descuento_porcentaje / 100
        self.precio -= descuento
        print(f"Se aplicó un descuento del {descuento_porcentaje}%. Precio actual: ${self.precio}")


````

````python
pantalon = Ropa("Pantalón", "M", "Negro", 200)
pantalon.info()
````

````python
pantalon.descuento(10)
````
